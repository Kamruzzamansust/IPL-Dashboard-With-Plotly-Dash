```python
import pandas as pd
from plotly.graph_objects import Figure, Scattermapbox

# Assuming your data frame (df) is loaded

# Define map center coordinates (longitude, latitude)
center_longitude = 77.208887  # Adjust these values as needed for Dhaka
center_latitude = 23.7104  # Adjust these values as needed for Dhaka

# Create a figure object
fig = Figure()

# Set mapbox access token (replace with your own token)
layout = dict(
    mapbox_style="open-street-map",
    accesstoken="YOUR_MAPBOX_ACCESS_TOKEN",  # Replace with your Mapbox access token
    center=dict(lon=center_longitude, lat=center_latitude),
    zoom=3,  # Adjust zoom level as desired
)

# Create Scattermapbox trace and add it to the figure
mapbox = Scattermapbox(
    lat=df['Latitude'],
    lon=df['Longitude'],
    marker_size=10,  # Adjust marker size as desired
    marker_color='blue',  # Adjust marker color as desired
    opacity=0.7,  # Adjust marker opacity as desired
)
fig.add_trace(mapbox)

# Set the layout for the figure
fig.update_layout(layout)

# Display the map
fig.show()
```

```python
import pandas as pd

# Assuming your dataset is stored in a variable named 'matches_df'
# Load the dataset into a pandas DataFrame
#matches_df = pd.read_csv("your_dataset.csv")  # Replace "your_dataset.csv" with the actual path to your dataset

# Standardize team names
df2['team1'] = df2['team1'].replace({'Rising Pune Supergiants': 'Rising Pune Supergiant'})
df2['team2'] = df2['team2'].replace({'Rising Pune Supergiants': 'Rising Pune Supergiant'})

# Count the number of matches played by each team using unique match IDs
matches_played_by_team = pd.concat([df2.groupby('team1')['id'].nunique(), 
                                    df2.groupby('team2')['id'].nunique()]).groupby(level=0).sum()

print("Number of matches played by each team:")
matches_played_by_team = matches_played_by_team.reset_index()
matches_played_by_team = matches_played_by_team.rename(columns={'index': 'Team_name', 'id': 'match_played'})
matches_played_by_team['match_played'].sum()
```
