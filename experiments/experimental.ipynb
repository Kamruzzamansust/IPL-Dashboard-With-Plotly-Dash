```python
import pandas as pd

# Assuming df1 is your DataFrame containing the data
def show_run(df, matchid, batsmanname):
    x = df[df['match_id'] == matchid][['batter', 'batsman_runs', 'over', 'ball']]
    x = x[x['batter'] == batsmanname]['batsman_runs'].sum()
    return x

# List of match IDs
match_ids = df1['match_id'].unique() # Add more match IDs as needed

# Create an empty list to store tuples of batsman names and their runs
all_batsman_runs = []

# Iterate over each match ID
for match_id in match_ids:
    # Get unique batsman names for the current match ID
    unique_batsmen = df1[df1['match_id'] == match_id]['batter'].unique()
    
    # Iterate over unique batsmen for the current match
    for batsmanname in unique_batsmen:
        runs = show_run(df1, match_id, batsmanname)
        all_batsman_runs.append((match_id, batsmanname, runs))

# Convert the list of tuples into a DataFrame
batsman_name_and_runs = pd.DataFrame(all_batsman_runs, columns=['match_id', 'batsman', 'runs'])

print(batsman_name_and_runs)
```

```python
batsman_name_and_runs.groupby(by='batsman')['runs'].sum().reset_index().sort_values(by='runs',ascending=False).head(10)
```

Now its time to get the number how many balls faced by each batsman



```python
import pandas as pd

# Assuming df1 is your DataFrame containing the data
def show_ball(df, matchid, batsmanname):
    x = df[df['match_id'] == matchid][['batter', 'batsman_runs', 'over', 'ball']]
    x = x[x['batter'] == batsmanname]['ball'].count()
    return x

# List of match IDs
match_ids = df1['match_id'].unique() # Add more match IDs as needed

# Create an empty list to store tuples of batsman names and their runs
all_batsman_balls = []

# Iterate over each match ID
for match_id in match_ids:
    # Get unique batsman names for the current match ID
    unique_batsmen = df1[df1['match_id'] == match_id]['batter'].unique()
    
    # Iterate over unique batsmen for the current match
    for batsmanname in unique_batsmen:
        balls = show_ball(df1, match_id, batsmanname)
        all_batsman_balls.append((match_id, batsmanname, balls))

# Convert the list of tuples into a DataFrame
batsman_name_and_balls = pd.DataFrame(all_batsman_balls, columns=['match_id', 'batsman', 'balls_faced'])

print(batsman_name_and_balls)
```

```python
batsman_name_and_balls.groupby(by='batsman')['balls_faced'].sum().reset_index().sort_values(by='balls_faced',ascending=False).head(10)
```

```python
import pandas as pd

# Assuming batsman_name_and_runs and batsman_name_and_balls are your DataFrames

# Merge the two DataFrames on the 'batsman' column
X = pd.merge(batsman_name_and_balls.groupby(by='batsman')['balls_faced'].sum().reset_index().sort_values(by='balls_faced',ascending=False), batsman_name_and_runs.groupby(by='batsman')['runs'].sum().reset_index().sort_values(by='runs',ascending=False), on='batsman')

X['strike_rate'] = round((X['runs']/X['balls_faced']) * 100,2)
```
